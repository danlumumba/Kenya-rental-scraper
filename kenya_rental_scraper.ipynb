```python
# import modules
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
current_page = 1
data = []
proceed  = True

while(proceed):
    print("currently scrapping page: " +str(current_page))
    url = f"https://www.property24.co.ke/houses-to-rent?page={current_page}"
    page = requests.get(url)
    soup = BeautifulSoup(page.text, "html.parser")
    
    #print(soup.prettify())

    if soup.title.text == "404 Not Found":
         proceed = False
         break
    else:
         all_rentals = soup.find_all('div', class_ ='col-xs-9')
         print(f"Found {len(all_rentals)} listings on page {current_page}")
         if not all_rentals:
               proceed = False
               break
         
         for rentals in all_rentals:
             if rentals.find('span', class_='p24_propertyTitle'):
               item = {}
               item['Title'] = rentals.find('span', class_='p24_propertyTitle').text.strip() if rentals.find('span', class_='p24_propertyTitle') else None
               item['Location'] = rentals.find('span', class_= 'p24_location').text.strip() if rentals.find('span', class_='p24_location') else None
               item['Bedrooms'] = rentals.find('span', title ='Bedrooms').text.strip() if rentals.find('span', title ='Bedrooms') else None
               item['Bathrooms'] = rentals.find('span', title ='Bathrooms').text.strip() if rentals.find('span', title ='Bathrooms') else None
               item['House Size'] = rentals.find('span', class_='p24_size').text.strip() if rentals.find('span', class_='p24_size') else None
               item['Description'] = rentals.find('span', class_='p24_excerpt').text.strip() if rentals.find('span', class_='p24_excerpt') else None
               item['Price'] = rentals.find('span', class_='p24_price').text.strip() if rentals.find('span', class_='p24_price') else None
               data.append(item)

         current_page += 1
# Convert to DataFrame, drop fully empty rows, and save to CSV
df = pd.DataFrame(data)
df.dropna(how='all', inplace=True)
df.to_csv("kenyarentalsfile1.csv", index = False)
print("Scraping complete. Data saved to 'kenyarentalsfile1.csv'")


```
